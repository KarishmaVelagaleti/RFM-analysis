### RFM Analysis

```python
rfm = RevCustomers[["customer id ", "date", "Revenue generated"]]
```

```python
rfm.head()
```

```python
rfm.rename(columns={"date":"Purchase Date"}, inplace=True)
```

```python
rfm.rename(columns={"customer id ":"Customer ID"}, inplace=True)
```

```python
rfm.head()
```

```python
rf = pd.pivot_table(data=rfm, index="Customer ID", values="Purchase Date",  aggfunc={'Purchase Date':[max, 'count']}).reset_index()
```

```python
rf["recency"] = (pd.to_datetime('2018-08-01')-(rf["max"])).dt.days+1
```

```python
rf.head()
```

```python
rf.info()
```

```python
rf.rename(columns={'count':'frequency'}, inplace=True)
```

```python
rf.head()
```

```python
m = pd.pivot_table(rfm, index='Customer ID', values="Revenue generated", aggfunc=np.sum).reset_index()
```

```python
m.head()
```

```python
RFM = pd.merge(rf, m, how='inner', on="Customer ID")
RFM.head()
```

```python
RFM.rename(columns={'Revenue generated':'monetary'}, inplace=True)
```

```python
RFM.describe()
```

```python
RFM["R"]=pd.qcut(RFM['recency'],q=[0,.25,0.75,1], labels=[1,2,3])
```

```python
RFM["F"]=pd.qcut(RFM['frequency'],q=[0,.25,0.75,1],labels=[2,1],duplicates='drop')
```

```python
RFM["M"]=pd.qcut(RFM['monetary'],q=[0,.25,0.75,1],labels=[3,2,1],duplicates='drop')
```

```python
RFM.head()
```

```python
RFM["R"]=RFM["R"].astype('int64')
```

```python
RFM["F"]=RFM["F"].astype('int64')
RFM["M"]=RFM["M"].astype('int64')
```

```python
RFM["RFM Score"]= RFM["R"].map(str)+RFM["F"].map(str)+RFM["M"].map(str)
```

```python
RFM.head()
```

```python
RFM.drop(columns=['max'], inplace=True)
```
